```python
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, WhiteKernel
from scipy.stats import norm, qmc

import matplotlib.pyplot as plt


###############################################################################
#                           TestFunction Class                                #
###############################################################################

class TestFunction:
    def __init__(self, func_type, n_x, track_x=False, x_shift=None, bounds=None):
        self.f_list = []
        self.x_list = []
        self.best_f = []

        self.func_type = func_type
        self.n_x = n_x
        self.track_x = track_x

        if x_shift is not None:
            self.x_shift = x_shift
        else:
            self.x_shift = np.zeros((n_x, 1))

        if bounds is not None:
            self.lb = bounds[:, 0]
            self.ub = bounds[:, 1]
        else:
            self.lb = np.zeros(n_x)
            self.ub = np.ones(n_x)

        self.dim = n_x
        self.int_var = np.array([], dtype=int)
        self.cont_var = np.arange(0, n_x)

    def eval(self, x):
        x_arr = np.array(x).reshape((self.n_x, 1)) + self.x_shift

        if self.func_type == "Rosenbrock_f":
            val = ((1.0 - x_arr) ** 2).sum() + \
                  100.0 * ((x_arr[1:] - x_arr[:-1] ** 2) ** 2).sum()

        elif self.func_type == "Levy_f":
            w = 1.0 + (x_arr - 1.0) / 4.0
            term1 = np.sin(np.pi * w[0]) ** 2
            term2 = ((w[:-1] - 1.0) ** 2 *
                     (1.0 + 10.0 * np.sin(np.pi * w[:-1] + 1.0) ** 2)).sum()
            term3 = (w[-1] - 1.0) ** 2 * \
                    (1.0 + np.sin(2.0 * np.pi * w[-1]) ** 2)
            val = float(term1 + term2 + term3)

        elif self.func_type == "Rastrigin_f":
            val = 10.0 * self.n_x + \
                  (x_arr ** 2 - 10.0 * np.cos(2.0 * np.pi * x_arr)).sum()
            val = float(val)

        elif self.func_type == "Ackley_f":
            a, b, c = 20.0, 0.2, 2.0 * np.pi
            norm_sq = (x_arr ** 2).sum()
            term1 = -a * np.exp(-b * np.sqrt(norm_sq / self.n_x))
            term2 = -np.exp(np.cos(c * x_arr).sum() / self.n_x)
            val = float(term1 + term2 + a + np.e)

        elif self.func_type == "1_norm":
            val = float(np.abs(x_arr).sum())

        else:
            raise ValueError(f"Unsupported function type: {self.func_type}")

        self.f_list.append(val)
        if self.track_x:
            self.x_list.append(x_arr.copy())
        return val


###############################################################################
#                         0) Repro and device                                 #
###############################################################################

SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")


###############################################################################
#             Lok style benchmark construction (Appendix A.3 style)           #
###############################################################################

dim = 20          # ambient dimension
intrinsic_dim = 5 # true dimensionality

bounds_z = np.array([[-5.0, 5.0]] * intrinsic_dim)

# choose the test function here
# options: "Rosenbrock_f", "Ackley_f", "Levy_f", "Rastrigin_f", "1_norm"
func_name = "Rastrigin_f"

t_func = TestFunction(
    func_type=func_name,
    n_x=intrinsic_dim,
    track_x=False,
    x_shift=None,
    bounds=bounds_z,
)

# structured manifold data
N = 10000
sampler = qmc.Sobol(d=intrinsic_dim, scramble=True)
Z = qmc.scale(
    sampler.random(N),
    l_bounds=bounds_z[:, 0],
    u_bounds=bounds_z[:, 1],
)  # (N, intrinsic_dim)

# scale that controls how "small" the true coords look in
scale_z = 0.5
G1 = scale_z * np.sinh(Z)   # monotone but nonlinear in Z

# group 2: no linear redundant dims
lin_dim = 0
G2 = np.zeros((N, 0), dtype=np.float32)

# group 3: big, highly nonlinear redundant dims
nonlin_dim = 12
A2 = np.random.randn(intrinsic_dim, nonlin_dim) / np.sqrt(intrinsic_dim)
H = Z @ A2
G3 = 5.0 * (np.tanh(H) + 0.2 * np.sin(2.0 * H) + 0.1 * H**2)

# group 4: large *heteroscedastic* correlated noise
corr_noise_dim = 3
A3 = np.random.randn(intrinsic_dim, corr_noise_dim) / np.sqrt(intrinsic_dim)

# norm-based scale: points further from origin in Z have larger noise variance
normZ = np.linalg.norm(Z, axis=1, keepdims=True) / np.sqrt(intrinsic_dim)
sigma_corr_base = 0.5 + 0.5 * normZ

# correlated part + independent jitter, then scaled per-sample
eps_corr = Z @ A3 + 0.3 * np.random.randn(N, corr_noise_dim)
G4 = sigma_corr_base * eps_corr

# group 5: pure high-variance noise
pure_noise_dim = dim - (intrinsic_dim + lin_dim + nonlin_dim + corr_noise_dim)
sigma_pure = 2.0

if pure_noise_dim > 0:
    G5 = sigma_pure * np.random.randn(N, pure_noise_dim)
    X = np.concatenate([G1, G2, G3, G4, G5], axis=1).astype(np.float32)
else:
    X = np.concatenate([G1, G2, G3, G4], axis=1).astype(np.float32)

# measurement jitter (small)
X = X + 0.01 * np.random.randn(*X.shape)

lb = X.min(axis=0)
ub = X.max(axis=0)

global_bounds_20d = np.stack([lb, ub], axis=1)


def recover_z_from_x(x):
    """
    Map from first intrinsic_dim coordinates of X back to latent Z
    using the inverse of G1 = scale_z * sinh(Z).
    """
    x = np.asarray(x, dtype=np.float32)
    x_first = x[:intrinsic_dim]  # shape (intrinsic_dim,)
    z_est = np.arcsinh(x_first / scale_z)
    return z_est.reshape(intrinsic_dim, 1)


###############################################################################
#                         2) AE and VAE model                                 #
###############################################################################

LATENT_DIM = 5


class VAE(nn.Module):
    def __init__(self, input_dim=20, latent_dim=LATENT_DIM, mean_decoder=False):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Linear(input_dim, 128), nn.SiLU(),
            nn.Linear(128, 64), nn.SiLU(),
            nn.Linear(64, 32), nn.SiLU(),
        )
        self.mu = nn.Linear(32, latent_dim)
        self.logvar = nn.Linear(32, latent_dim)

        self.dec = nn.Sequential(
            nn.Linear(latent_dim, 32), nn.SiLU(),
            nn.Linear(32, 64), nn.SiLU(),
            nn.Linear(64, 128), nn.SiLU(),
            nn.Linear(128, input_dim),
        )

        self.mean_decoder = mean_decoder

    def reparam(self, mu, lv):
        std = torch.exp(0.5 * lv)
        eps = torch.randn_like(std)
        return mu + eps * std

    def forward(self, x):
        h = self.enc(x)
        mu = self.mu(h)
        lv = self.logvar(h)

        if self.mean_decoder:
            z = mu
        else:
            z = self.reparam(mu, lv)

        xhat = self.dec(z)
        return xhat, mu, lv, z

    @torch.no_grad()
    def encode_mu(self, x):
        h = self.enc(x)
        return self.mu(h)

    @torch.no_grad()
    def decode(self, z):
        return self.dec(z)


###############################################################################
#                 3) Shared trainer for AE and VAE                            #
###############################################################################

def train_autoencoder(
    X_full,
    latent_dim=LATENT_DIM,
    lr=0.002071334578703477,
    batch_size=128,
    max_epochs=800,
    patience=100,
    seed=42,
    kind="vae",               # "ae" or "vae"
    beta_start=0.0,
    beta_inc=0.01,
    beta_max=1.0,
    beta_freeze_epochs=20,
    log_every=20,
):
    assert kind in ["ae", "vae"]

    torch.manual_seed(seed)
    np.random.seed(seed)

    X_full = np.asarray(X_full, dtype=np.float32)
    N, D = X_full.shape

    idx = np.random.permutation(N)
    split = int(0.8 * N)
    tr_idx, va_idx = idx[:split], idx[split:]

    Xtr_raw, Xva_raw = X_full[tr_idx], X_full[va_idx]

    mu = Xtr_raw.mean(axis=0, keepdims=True).astype(np.float32)
    std = (Xtr_raw.std(axis=0, keepdims=True) + 1e-8).astype(np.float32)

    Xtr = (Xtr_raw - mu) / std
    Xva = (Xva_raw - mu) / std

    Xtr = np.ascontiguousarray(Xtr, dtype=np.float32)
    Xva = np.ascontiguousarray(Xva, dtype=np.float32)

    train_loader = DataLoader(
        TensorDataset(torch.from_numpy(Xtr), torch.from_numpy(Xtr)),
        batch_size=batch_size,
        shuffle=True,
        drop_last=False,
    )
    val_loader = DataLoader(
        TensorDataset(torch.from_numpy(Xva), torch.from_numpy(Xva)),
        batch_size=256,
        shuffle=False,
        drop_last=False,
    )

    use_mean_decoder = (kind == "ae")
    model = VAE(input_dim=D, latent_dim=latent_dim,
                mean_decoder=use_mean_decoder).to(device)

    opt = Adam(model.parameters(), lr=lr)
    sched = ReduceLROnPlateau(opt, mode="min", factor=0.5,
                              patience=20, min_lr=1e-6)
    mse = nn.MSELoss(reduction="mean")

    train_hist, val_hist, beta_hist, kl_hist = [], [], [], []
    best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
    best_val_recon = np.inf
    no_imp = 0

    beta = beta_start

    for epoch in range(1, max_epochs + 1):
        if kind == "ae":
            beta_eff = 0.0
        else:
            if epoch <= beta_freeze_epochs:
                beta_eff = 0.0
            else:
                beta = min(beta + beta_inc, beta_max)
                beta_eff = beta

        model.train()
        tr_sum = 0.0
        kl_sum = 0.0
        n_tr = 0

        for xb, _ in train_loader:
            xb = xb.to(device)
            opt.zero_grad()
            xhat, mu_t, lv_t, _ = model(xb)
            recon = mse(xhat, xb)

            if kind == "vae" and beta_eff > 0.0:
                kl = -0.5 * torch.mean(
                    torch.sum(1 + lv_t - mu_t.pow(2) - lv_t.exp(), dim=1)
                )
            else:
                kl = torch.tensor(0.0, device=xb.device)

            loss = recon + beta_eff * kl
            loss.backward()
            opt.step()

            tr_sum += loss.item() * xb.size(0)
            kl_sum += kl.item() * xb.size(0)
            n_tr += xb.size(0)

        tr_loss = tr_sum / n_tr
        kl_avg = kl_sum / n_tr

        model.eval()
        rec_sum = 0.0
        n_va = 0
        with torch.no_grad():
            for xb, _ in val_loader:
                xb = xb.to(device)
                xhat, mu_t, lv_t, _ = model(xb)
                recon = mse(xhat, xb)
                rec_sum += recon.item() * xb.size(0)
                n_va += xb.size(0)
        val_recon = rec_sum / n_va

        train_hist.append(tr_loss)
        val_hist.append(val_recon)
        beta_hist.append(beta_eff)
        kl_hist.append(kl_avg)

        sched.step(val_recon)

        if epoch == 1 or epoch % log_every == 0:
            print(
                f"[{kind.upper()}] Ep {epoch:4d}  beta {beta_eff:.3f}  "
                f"Train {tr_loss:.4f}  ValRecon {val_recon:.4f}  KL {kl_avg:.4f}"
            )

        if val_recon < best_val_recon - 1e-12:
            best_val_recon = val_recon
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            no_imp = 0
        else:
            no_imp += 1
            if no_imp >= patience:
                print(f"[{kind.upper()}] early stop at epoch {epoch}")
                break

    model.load_state_dict(best_state)
    model.to(device)
    model.eval()

    with torch.no_grad():
        Xt = torch.from_numpy(Xva).to(device)
        xhat = model(Xt)[0]
        final_mse = float(((xhat - Xt) ** 2).mean().cpu().item())

    print(f"[{kind.upper()}] best validation recon MSE: {final_mse:.6f}")

    history = {
        "train": np.array(train_hist),
        "val": np.array(val_hist),
        "beta": np.array(beta_hist),
        "kl": np.array(kl_hist),
    }
    stats = {
        "mu": mu,
        "std": std,
        "val_mse_best": final_mse,
    }
    return model, history, stats


###############################################################################
#                         4) GP, EI and BO utilities                          #
###############################################################################

def make_gp(latent_dim):
    kernel = 1.0 * Matern(
        length_scale=np.ones(latent_dim),
        length_scale_bounds=(1e-2, 1e2),
        nu=2.5,
    )
    kernel += WhiteKernel(
        noise_level=1e-6,
        noise_level_bounds=(1e-8, 1e-2),
    )

    return GaussianProcessRegressor(
        kernel=kernel,
        n_restarts_optimizer=5,
        alpha=1e-6,
        normalize_y=True,
    )


def expected_improvement(x, gp, best_y, xi):
    x = np.atleast_2d(x)
    mu, sigma = gp.predict(x, return_std=True)
    sigma = sigma + 1e-9
    imp = best_y - mu - xi
    z = imp / sigma
    ei = imp * norm.cdf(z) + sigma * norm.pdf(z)
    return ei.ravel()


def propose_sobol_local(
    gp,
    best_y,
    center,
    radius,
    global_bounds,
    latent_dim,
    xi,
    n_candidates=512,
):
    lo_g = global_bounds[:, 0]
    hi_g = global_bounds[:, 1]

    lo = np.maximum(center - radius, lo_g)
    hi = np.minimum(center + radius, hi_g)

    span = hi - lo
    span[span < 1e-6] = 1e-6
    lo = center - 0.5 * span
    hi = center + 0.5 * span

    sampler = qmc.Sobol(d=latent_dim, scramble=True)
    U = sampler.random(n_candidates)
    Zc = qmc.scale(U, l_bounds=lo, u_bounds=hi)

    ei = expected_improvement(Zc, gp, best_y, xi)
    idx = np.argmax(ei)
    return Zc[idx]


def optimise_latent_BO(
    latent_dim,
    objective,
    global_bounds,
    n_initial=40,
    n_iter=40,
    n_candidates=512,
    init_radius=1.0,
    min_radius=0.1,
    seed=42,
    verbose=False,
):
    rng = np.random.RandomState(seed)

    lo = global_bounds[:, 0]
    hi = global_bounds[:, 1]

    sampler = qmc.Sobol(d=latent_dim, scramble=True)
    U0 = sampler.random(n_initial)
    Z = qmc.scale(U0, l_bounds=lo, u_bounds=hi)

    y = np.array([objective(z) for z in Z])

    best_idx = np.argmin(y)
    best_z = Z[best_idx].copy()
    best_f = y[best_idx]
    radius = init_radius

    gp = make_gp(latent_dim)
    history = [best_f]

    full_span = hi - lo

    for it in range(n_iter):
        gp.fit(Z, y)
        best_y = y.min()

        frac = it / max(1, n_iter - 1)
        xi = 0.1 * (1.0 - frac) + 0.01 * frac

        z_new = propose_sobol_local(
            gp=gp,
            best_y=best_y,
            center=best_z,
            radius=radius,
            global_bounds=global_bounds,
            latent_dim=latent_dim,
            xi=xi,
            n_candidates=n_candidates,
        )

        z_new = z_new + 1e-4 * rng.randn(latent_dim)
        y_new = objective(z_new)

        Z = np.vstack((Z, z_new))
        y = np.append(y, y_new)

        if y_new < best_f:
            best_f = y_new
            best_z = z_new
            radius = min(radius * 1.1, full_span.max() / 2.0)
        else:
            radius = max(radius * 0.9, min_radius)

        history.append(best_f)

    if verbose:
        print(f"[BO] finished | best {best_f:.4e}")

    return best_z, best_f, np.array(history), Z, y


###############################################################################
#                          SDR helpers                                        #
###############################################################################

def propose_sobol_in_box(
    gp,
    best_y,
    lo,
    hi,
    latent_dim,
    xi,
    n_candidates=512,
    rng=None,
):
    lo = np.asarray(lo, dtype=np.float64)
    hi = np.asarray(hi, dtype=np.float64)

    sampler = qmc.Sobol(
        d=latent_dim,
        scramble=True,
        seed=None if rng is None else int(rng.randint(0, 10**9)),
    )
    U = sampler.random(n_candidates)
    Zc = qmc.scale(U, l_bounds=lo, u_bounds=hi)

    ei = expected_improvement(Zc, gp, best_y, xi)
    idx = np.argmax(ei)
    return Zc[idx]


def sdr_update_roi(
    best_z_history,
    roi_lo_prev,
    roi_hi_prev,
    global_lo,
    global_hi,
    r_prev,
    r_prev_prev,
    gamma0=0.7,
    gamma_p=1.0,
    eta=0.9,
    min_rel_width=0.02,
    r0=None,
):
    global_lo = np.asarray(global_lo, dtype=np.float64)
    global_hi = np.asarray(global_hi, dtype=np.float64)

    roi_lo_prev = np.asarray(roi_lo_prev, dtype=np.float64)
    roi_hi_prev = np.asarray(roi_hi_prev, dtype=np.float64)
    r_prev = np.asarray(r_prev, dtype=np.float64)

    if r0 is None:
        r0 = global_hi - global_lo
    r0 = np.asarray(r0, dtype=np.float64)

    r_min_abs = np.maximum(min_rel_width * r0, 1e-8)

    x_k = np.asarray(best_z_history[-1], dtype=np.float64)

    if len(best_z_history) < 3 or r_prev_prev is None:
        r_new = np.maximum(r_prev, r_min_abs)
        center = x_k
    else:
        x_k1 = np.asarray(best_z_history[-2], dtype=np.float64)
        x_k2 = np.asarray(best_z_history[-3], dtype=np.float64)

        r_prev_prev = np.asarray(r_prev_prev, dtype=np.float64)
        r_prev = np.maximum(r_prev, r_min_abs)
        r_prev_prev = np.maximum(r_prev_prev, r_min_abs)

        d_k = 2.0 * (x_k - x_k1) / r_prev
        d_k1 = 2.0 * (x_k1 - x_k2) / r_prev_prev

        c = d_k * d_k1
        c_hat = np.sign(c) * np.sqrt(np.abs(c))

        gamma = 0.5 * (gamma_p * (1.0 + c_hat) + gamma0 * (1.0 - c_hat))
        lam = eta + np.abs(d_k) * (gamma - eta)

        r_new = lam * r_prev
        r_new = np.maximum(r_new, r_min_abs)

        center = x_k

    roi_lo = np.maximum(center - 0.5 * r_new, global_lo)
    roi_hi = np.minimum(center + 0.5 * r_new, global_hi)

    r_eff = roi_hi - roi_lo
    r_eff = np.maximum(r_eff, r_min_abs)

    return roi_lo, roi_hi, r_eff, center


def optimise_latent_BO_sdr(
    latent_dim,
    objective,
    global_bounds,
    n_initial=40,
    n_iter=40,
    n_candidates=512,
    seed=42,
    gamma0=0.7,
    gamma_p=1.0,
    eta=0.9,
    min_rel_width=0.02,
    xi_update=1,
    t_min=None,
    verbose=False,
):
    rng = np.random.RandomState(seed)

    global_bounds = np.asarray(global_bounds, dtype=np.float64)
    lo_g = global_bounds[:, 0]
    hi_g = global_bounds[:, 1]

    r0 = hi_g - lo_g

    sampler = qmc.Sobol(d=latent_dim, scramble=True,
                        seed=int(rng.randint(0, 10**9)))
    U0 = sampler.random(n_initial)
    Z = qmc.scale(U0, l_bounds=lo_g, u_bounds=hi_g)

    y = np.array([objective(z) for z in Z])

    best_idx = np.argmin(y)
    best_z = Z[best_idx].copy()
    best_f = y[best_idx]

    center0 = best_z.copy()
    roi_lo = np.maximum(center0 - 0.5 * r0, lo_g)
    roi_hi = np.minimum(center0 + 0.5 * r0, hi_g)
    r_curr = roi_hi - roi_lo

    gp = make_gp(latent_dim)
    history = [best_f]
    best_z_history = [best_z.copy()]
    r_history = [r_curr.copy()]

    for it in range(n_iter):
        gp.fit(Z, y)
        best_y = y.min()

        frac = it / max(1, n_iter - 1)
        xi_EI = 0.1 * (1.0 - frac) + 0.01 * frac

        z_new = propose_sobol_in_box(
            gp=gp,
            best_y=best_y,
            lo=roi_lo,
            hi=roi_hi,
            latent_dim=latent_dim,
            xi=xi_EI,
            n_candidates=n_candidates,
            rng=rng,
        )
        z_new = z_new + 1e-4 * rng.randn(latent_dim)

        y_new = objective(z_new)

        Z = np.vstack((Z, z_new))
        y = np.append(y, y_new)

        if y_new < best_f:
            best_f = y_new
            best_z = z_new

        best_z_history.append(best_z.copy())
        history.append(best_f)

        do_update = (xi_update is not None) and ((it + 1) % xi_update == 0)

        if do_update:
            r_prev = r_history[-1]
            r_prev_prev = r_history[-2] if len(r_history) >= 2 else None

            if t_min is not None and np.all(r_prev <= t_min):
                r_history.append(r_prev.copy())
                continue

            roi_lo, roi_hi, r_curr, _ = sdr_update_roi(
                best_z_history=best_z_history,
                roi_lo_prev=roi_lo,
                roi_hi_prev=roi_hi,
                global_lo=lo_g,
                global_hi=hi_g,
                r_prev=r_prev,
                r_prev_prev=r_prev_prev,
                gamma0=gamma0,
                gamma_p=gamma_p,
                eta=eta,
                min_rel_width=min_rel_width,
                r0=r0,
            )
            r_history.append(r_curr.copy())
        else:
            r_history.append(r_curr.copy())

    if verbose:
        print(f"[BO SDR] finished | best {best_f:.4e}")

    return best_z, best_f, np.array(history), Z, y


###############################################################################
#                  5) Latent components for BO for a model                    #
###############################################################################

def build_latent_bo_components(
    model,
    mu_stats,
    std_stats,
    X,
    lb,
    ub,
    label,
    t_func,
    intrinsic_dim,
):
    print(f"\nBuilding latent space for {label}")

    with torch.no_grad():
        Xn = ((X - mu_stats) / std_stats).astype(np.float32)
        Z_full = model.encode_mu(torch.from_numpy(Xn).to(device)).cpu().numpy()

    scZ = StandardScaler().fit(Z_full)
    Z_wh = scZ.transform(Z_full)

    latent_dim = Z_wh.shape[1]
    z_lo = np.quantile(Z_wh, 0.01, axis=0)
    z_hi = np.quantile(Z_wh, 0.99, axis=0)
    global_bounds = np.stack([z_lo, z_hi], axis=1)

    @torch.no_grad()
    def decode_from_whitened(z_wh_np):
        z = scZ.inverse_transform(
            np.asarray(z_wh_np, dtype=np.float32)[None, :]
        )
        z_t = torch.from_numpy(z.astype(np.float32)).to(device)
        xhat_norm = model.decode(z_t).cpu().numpy()
        xhat = xhat_norm * std_stats + mu_stats
        xhat = np.clip(xhat[0], lb, ub)
        return xhat

    def latent_objective(z_wh):
        x_dec = decode_from_whitened(z_wh)  # 1D array length dim
        z_est = recover_z_from_x(x_dec)     # shape (intrinsic_dim, 1)
        return float(t_func.eval(z_est))

    components = {
        "latent_dim": latent_dim,
        "global_bounds": global_bounds,
        "decode_fn": decode_from_whitened,
        "objective": latent_objective,
        "scaler": scZ,
    }
    return components


###############################################################################
#                  5b) PCA latent components for BO                           #
###############################################################################

def build_pca_bo_components(
    X,
    lb,
    ub,
    t_func,
    intrinsic_dim,
    latent_dim_pca=5,
):
    print("\nBuilding latent space for PCA")

    X = np.asarray(X, dtype=np.float32)

    # centre, but do NOT feature-wise scale
    X_mean = X.mean(axis=0, keepdims=True)
    Xc = X - X_mean

    pca = PCA(n_components=latent_dim_pca)
    Z_full = pca.fit_transform(Xc)  # (N, latent_dim_pca)

    # whiten latent for GP convenience
    scZ = StandardScaler().fit(Z_full)
    Z_wh = scZ.transform(Z_full)

    z_lo = np.quantile(Z_wh, 0.01, axis=0)
    z_hi = np.quantile(Z_wh, 0.99, axis=0)
    global_bounds = np.stack([z_lo, z_hi], axis=1)

    def decode_from_whitened_pca(z_wh_np):
        z_wh_np = np.asarray(z_wh_np, dtype=np.float32).reshape(1, -1)
        z_lat = scZ.inverse_transform(z_wh_np)
        Xc_rec = pca.inverse_transform(z_lat)
        x_rec = Xc_rec + X_mean
        x_rec = np.clip(x_rec[0], lb, ub)
        return x_rec

    def latent_objective_pca(z_wh):
        x_dec = decode_from_whitened_pca(z_wh)  # 1D array length dim
        z_est = recover_z_from_x(x_dec)         # shape (intrinsic_dim, 1)
        return float(t_func.eval(z_est))

    components = {
        "latent_dim": latent_dim_pca,
        "global_bounds": global_bounds,
        "decode_fn": decode_from_whitened_pca,
        "objective": latent_objective_pca,
        "X_mean": X_mean,
        "scaler_Z": scZ,
        "pca": pca,
    }
    return components


###############################################################################
#                         6) Train AE and VAE                                 #
###############################################################################

ae_model, ae_hist, ae_stats = train_autoencoder(
    X_full=X,
    latent_dim=LATENT_DIM,
    lr=1.5e-3,
    batch_size=128,
    max_epochs=400,
    patience=60,
    seed=SEED,
    kind="ae",
)

vae_model, vae_hist, vae_stats = train_autoencoder(
    X_full=X,
    latent_dim=LATENT_DIM,
    lr=1.5e-3,
    batch_size=128,
    max_epochs=600,
    patience=80,
    seed=SEED,
    kind="vae",
    beta_start=0.0,
    beta_inc=0.001,
    beta_max=0.05,
    beta_freeze_epochs=100,
)

print("\nAE best val MSE:", ae_stats["val_mse_best"])
print("VAE best val MSE:", vae_stats["val_mse_best"])


###############################################################################
#                         7) Optional: plot AE and VAE losses                 #
###############################################################################

plt.figure(figsize=(6, 4))
plt.plot(ae_hist["train"], label="AE train")
plt.plot(ae_hist["val"], label="AE val")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("AE training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure(figsize=(6, 4))
plt.plot(vae_hist["train"], label="VAE train")
plt.plot(vae_hist["val"], label="VAE val")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.title("VAE training and validation loss")
plt.legend()
plt.tight_layout()
plt.show()


###############################################################################
#                         8) Build VAE latent space                           #
###############################################################################

vae_components = build_latent_bo_components(
    model=vae_model,
    mu_stats=vae_stats["mu"],
    std_stats=vae_stats["std"],
    X=X,
    lb=lb,
    ub=ub,
    label="VAE",
    t_func=t_func,
    intrinsic_dim=intrinsic_dim,
)


###############################################################################
#                         8b) Build PCA latent space                          #
###############################################################################

PCA_LATENT_DIM = 5
pca_components = build_pca_bo_components(
    X=X,
    lb=lb,
    ub=ub,
    t_func=t_func,
    intrinsic_dim=intrinsic_dim,
    latent_dim_pca=PCA_LATENT_DIM,
)


###############################################################################
#                9) Direct BO on full 20D function (baseline)                 #
###############################################################################

def objective_20d(x):
    x = np.asarray(x, dtype=np.float32)
    z_est = recover_z_from_x(x)  # use the same inverse warp
    return float(t_func.eval(z_est))  # t_func is intrinsic-dim test function


###############################################################################
#                     10) Settings for repeated BO experiments                #
###############################################################################

N_REP = 5
N_REP_PCA = 3  # PCA runs
N_INIT = 40
N_ITER = 100
N_CAND = 512
INIT_RAD = 1.0
MIN_RAD = 0.1


###############################################################################
# 11) Native BO in 20D (separate loop)                                        #
###############################################################################

hist_native_runs = []
best_native_final = []

print("\n=== Native BO in 20D x space ===")
for r in range(N_REP):
    seed_r = SEED + 100 * r
    _, best_f_native, hist_native, _, _ = optimise_latent_BO(
        latent_dim=dim,
        objective=objective_20d,
        global_bounds=global_bounds_20d,
        n_initial=N_INIT,
        n_iter=N_ITER,
        n_candidates=N_CAND,
        init_radius=INIT_RAD,
        min_radius=MIN_RAD,
        seed=seed_r,
        verbose=False,
    )
    hist_native_runs.append(hist_native)
    best_native_final.append(best_f_native)
    print(f"run {r+1}/{N_REP} native BO final best: {best_f_native:.4e}")

hist_native_runs = np.stack(hist_native_runs, axis=0)
best_native_final = np.array(best_native_final)


###############################################################################
# 12) BO in VAE latent space (separate loop)                                  #
###############################################################################

hist_vae_runs = []
best_vae_final = []

print("\n=== BO in VAE latent space ===")
for r in range(N_REP):
    seed_r = SEED + 200 * r
    _, best_f_vae, hist_vae, _, _ = optimise_latent_BO(
        latent_dim=vae_components["latent_dim"],
        objective=vae_components["objective"],
        global_bounds=vae_components["global_bounds"],
        n_initial=N_INIT,
        n_iter=N_ITER,
        n_candidates=N_CAND,
        init_radius=INIT_RAD,
        min_radius=MIN_RAD,
        seed=seed_r,
        verbose=False,
    )
    hist_vae_runs.append(hist_vae)
    best_vae_final.append(best_f_vae)
    print(f"run {r+1}/{N_REP} VAE BO final best: {best_f_vae:.4e}")

hist_vae_runs = np.stack(hist_vae_runs, axis=0)
best_vae_final = np.array(best_vae_final)


###############################################################################
# 13) BO plus SDR in VAE latent space (separate loop)                         #
###############################################################################

hist_vae_sdr_runs = []
best_vae_sdr_final = []

print("\n=== BO plus SDR in VAE latent space ===")
for r in range(N_REP):
    seed_r = SEED + 300 * r
    _, best_f_vae_sdr, hist_vae_sdr, _, _ = optimise_latent_BO_sdr(
        latent_dim=vae_components["latent_dim"],
        objective=vae_components["objective"],
        global_bounds=vae_components["global_bounds"],
        n_initial=N_INIT,
        n_iter=N_ITER,
        n_candidates=N_CAND,
        seed=seed_r,
        gamma0=0.7,
        gamma_p=1.0,
        eta=0.9,
        min_rel_width=0.02,
        xi_update=1,
        t_min=None,
        verbose=False,
    )
    hist_vae_sdr_runs.append(hist_vae_sdr)
    best_vae_sdr_final.append(best_f_vae_sdr)
    print(f"run {r+1}/{N_REP} VAE BO plus SDR final best: {best_f_vae_sdr:.4e}")

hist_vae_sdr_runs = np.stack(hist_vae_sdr_runs, axis=0)
best_vae_sdr_final = np.array(best_vae_sdr_final)


###############################################################################
# 13b) BO in PCA latent space (separate loop)                                 #
###############################################################################

hist_pca_runs = []
best_pca_final = []

print("\n=== BO in PCA latent space ===")
for r in range(N_REP_PCA):
    seed_r = SEED + 400 * r
    _, best_f_pca, hist_pca, _, _ = optimise_latent_BO(
        latent_dim=pca_components["latent_dim"],
        objective=pca_components["objective"],
        global_bounds=pca_components["global_bounds"],
        n_initial=N_INIT,
        n_iter=N_ITER,
        n_candidates=N_CAND,
        init_radius=INIT_RAD,
        min_radius=MIN_RAD,
        seed=seed_r,
        verbose=False,
    )
    hist_pca_runs.append(hist_pca)
    best_pca_final.append(best_f_pca)
    print(f"run {r+1}/{N_REP_PCA} PCA BO final best: {best_f_pca:.4e}")

hist_pca_runs = np.stack(hist_pca_runs, axis=0)
best_pca_final = np.array(best_pca_final)


###############################################################################
# 14) Plot optimality gap with bands (four methods together)                  #
###############################################################################

# global optimum f* = 0 for these test functions
f_star = 0.0

gap_native = hist_native_runs - f_star
gap_vae = hist_vae_runs - f_star
gap_vae_sdr = hist_vae_sdr_runs - f_star
gap_pca = hist_pca_runs - f_star


def mean_and_band(arr):
    mean = np.mean(arr, axis=0)
    lo = np.percentile(arr, 10, axis=0)
    hi = np.percentile(arr, 90, axis=0)
    return mean, lo, hi


mean_nat, lo_nat, hi_nat = mean_and_band(gap_native)
mean_vae, lo_vae, hi_vae = mean_and_band(gap_vae)
mean_vae_s, lo_vae_s, hi_vae_s = mean_and_band(gap_vae_sdr)
mean_pca, lo_pca, hi_pca = mean_and_band(gap_pca)

iters = np.arange(len(mean_nat))

plt.figure(figsize=(7, 4.5))
plt.semilogy(iters, mean_nat, color="k", label="BO on 20D")
plt.fill_between(iters, lo_nat, hi_nat, color="k", alpha=0.15)

plt.semilogy(iters, mean_vae, color="r", label="BO VAE latent")
plt.fill_between(iters, lo_vae, hi_vae, color="r", alpha=0.15)

plt.semilogy(iters, mean_vae_s, color="b", label="BO VAE SDR")
plt.fill_between(iters, lo_vae_s, hi_vae_s, color="b", alpha=0.15)

plt.semilogy(iters, mean_pca, color="g", label="BO PCA latent (5D)")
plt.fill_between(iters, lo_pca, hi_pca, color="g", alpha=0.15)

nice_name = func_name.replace("_f", "")
plt.xlabel("BO iteration")
plt.ylabel("optimality gap")
plt.xlim(0, 100)
plt.title(f"{nice_name} Function")
plt.legend()
plt.tight_layout()
plt.show()


###############################################################################
# 15) Final summary of best values                                            #
###############################################################################

print(
    "\nFinal best objective values over "
    f"{N_REP} runs (PCA uses {N_REP_PCA} runs), mean  std:"
)
print(
    f"BO native:           {best_native_final.mean():.4e} "
    f" {best_native_final.std():.4e}"
)
print(
    f"BO VAE latent:       {best_vae_final.mean():.4e} "
    f" {best_vae_final.std():.4e}"
)
print(
    f"BO VAE latent SDR:   {best_vae_sdr_final.mean():.4e} "
    f" {best_vae_sdr_final.std():.4e}"
)
print(
    f"BO PCA latent (5D):  {best_pca_final.mean():.4e} "
    f" {best_pca_final.std():.4e}"
)
```
