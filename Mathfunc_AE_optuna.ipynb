```python
import time
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import ReduceLROnPlateau

from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import Matern, WhiteKernel
from scipy.stats import norm
from scipy.stats import qmc

import matplotlib.pyplot as plt
import optuna

###############################################################################
#                           TestFunction Class                                #
###############################################################################

class TestFunction:
    def __init__(self, func_type, n_x, track_x=False, x_shift=None, bounds=None):
        self.f_list = []
        self.x_list = []
        self.best_f = []

        self.func_type = func_type
        self.n_x = n_x
        self.track_x = track_x

        if x_shift is not None:
            self.x_shift = x_shift
        else:
            self.x_shift = np.zeros((n_x, 1))
        if bounds is not None:
            self.lb = bounds[:, 0]
            self.ub = bounds[:, 1]
        else:
            self.lb = np.zeros(n_x)
            self.ub = np.ones(n_x)
        self.dim = n_x
        self.int_var = np.array([], dtype=int)
        self.cont_var = np.arange(0, n_x)

    def eval(self, x):
        x_arr = np.array(x).reshape((self.n_x, 1)) + self.x_shift

        if self.func_type == 'Rosenbrock_f':
            val = ((1.0 - x_arr) ** 2).sum() + 100.0 * ((x_arr[1:] - x_arr[:-1] ** 2) ** 2).sum()

        elif self.func_type == 'Levy_f':
            w = 1.0 + (x_arr - 1.0) / 4.0
            term1 = np.sin(np.pi * w[0]) ** 2
            term2 = ((w[:-1] - 1.0) ** 2 * (1.0 + 10.0 * np.sin(np.pi * w[:-1] + 1.0) ** 2)).sum()
            term3 = (w[-1] - 1.0) ** 2 * (1.0 + np.sin(2.0 * np.pi * w[-1]) ** 2)
            val = float(term1 + term2 + term3)

        elif self.func_type == 'Rastrigin_f':
            val = 10.0 * self.n_x + (x_arr ** 2 - 10.0 * np.cos(2.0 * np.pi * x_arr)).sum()
            val = float(val)

        elif self.func_type == 'Ackley_f':
            a, b, c = 20.0, 0.2, 2.0 * np.pi
            norm_sq = (x_arr ** 2).sum()
            term1 = -a * np.exp(-b * np.sqrt(norm_sq / self.n_x))
            term2 = -np.exp(np.cos(c * x_arr).sum() / self.n_x)
            val = float(term1 + term2 + a + np.e)

        elif self.func_type == '1_norm':
            val = float(np.abs(x_arr).sum())
        else:
            raise ValueError(f"Unsupported function type: {self.func_type}")

        self.f_list.append(val)
        if self.track_x:
            self.x_list.append(x_arr.copy())
        return val

###############################################################################
#                         0) Repro and device                                 #
###############################################################################

SEED = 42
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

###############################################################################
#                         1) Choose function and sample                       #
###############################################################################

dim = 20
func_name = "Ackley_f"
bounds = np.array([[-5.0, 5.0]] * dim)

t_func = TestFunction(
    func_type=func_name,
    n_x=dim,
    track_x=False,
    x_shift=None,
    bounds=bounds,
)

N = 10_000
lb = t_func.lb
ub = t_func.ub

U = np.random.rand(N, dim)
X = lb + U * (ub - lb)
X = X.astype(np.float32)

###############################################################################
#                      2) Bottleneck AE with width_mult                       #
###############################################################################

class BottleneckAE(nn.Module):
    """
    Encoder: 20 -> w1 -> w2 -> latent
    Decoder: latent -> w2 -> w1 -> 20
    w1, w2 are scaled by width_mult, giving a bottleneck shape.
    """

    def __init__(self, input_dim=20, latent_dim=5, width_mult=1.0):
        super().__init__()
        base_w1, base_w2 = 128, 64
        w1 = int(base_w1 * width_mult)
        w2 = int(base_w2 * width_mult)

        # encoder
        self.enc = nn.Sequential(
            nn.Linear(input_dim, w1),
            nn.LayerNorm(w1),
            nn.SiLU(),
            nn.Linear(w1, w2),
            nn.LayerNorm(w2),
            nn.SiLU(),
            nn.Linear(w2, latent_dim),
        )

        # decoder (mirror)
        self.dec = nn.Sequential(
            nn.Linear(latent_dim, w2),
            nn.SiLU(),
            nn.Linear(w2, w1),
            nn.LayerNorm(w1),
            nn.SiLU(),
            nn.Linear(w1, input_dim),
        )

    def encode_mu(self, x):
        return self.enc(x)

    def decode(self, z):
        return self.dec(z)

    def forward(self, x):
        z = self.encode_mu(x)
        xhat = self.decode(z)
        return xhat

###############################################################################
#                         3) AE training function                             #
###############################################################################

def train_ae(
    X_full,
    latent_dim,
    width_mult,
    lr,
    batch_size,
    max_epochs=400,
    patience=60,
    seed=42,
    log_every=50,
):
    torch.manual_seed(seed)
    np.random.seed(seed)

    X_full = np.asarray(X_full, dtype=np.float32)
    N, D = X_full.shape

    idx = np.random.permutation(N)
    split = int(0.8 * N)
    tr_idx, va_idx = idx[:split], idx[split:]

    Xtr_raw, Xva_raw = X_full[tr_idx], X_full[va_idx]

    mu = Xtr_raw.mean(axis=0, keepdims=True).astype(np.float32)
    std = (Xtr_raw.std(axis=0, keepdims=True) + 1e-8).astype(np.float32)

    Xtr = (Xtr_raw - mu) / std
    Xva = (Xva_raw - mu) / std

    train_loader = DataLoader(
        TensorDataset(torch.from_numpy(Xtr), torch.from_numpy(Xtr)),
        batch_size=batch_size,
        shuffle=True,
        drop_last=False,
    )
    val_loader = DataLoader(
        TensorDataset(torch.from_numpy(Xva), torch.from_numpy(Xva)),
        batch_size=256,
        shuffle=False,
        drop_last=False,
    )

    model = BottleneckAE(input_dim=D, latent_dim=latent_dim, width_mult=width_mult).to(device)

    opt = Adam(model.parameters(), lr=lr)
    sched = ReduceLROnPlateau(opt, mode="min", factor=0.5, patience=15, min_lr=1e-6)
    mse = nn.MSELoss(reduction="mean")

    best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
    best_val = np.inf
    no_imp = 0

    train_hist, val_hist = [], []

    for epoch in range(1, max_epochs + 1):
        model.train()
        tr_sum = 0.0
        n_tr = 0
        for xb, _ in train_loader:
            xb = xb.to(device)
            opt.zero_grad()
            xhat = model(xb)
            loss = mse(xhat, xb)
            loss.backward()
            opt.step()
            tr_sum += loss.item() * xb.size(0)
            n_tr += xb.size(0)
        tr_loss = tr_sum / n_tr

        model.eval()
        va_sum = 0.0
        n_va = 0
        with torch.no_grad():
            for xb, _ in val_loader:
                xb = xb.to(device)
                xhat = model(xb)
                loss = mse(xhat, xb)
                va_sum += loss.item() * xb.size(0)
                n_va += xb.size(0)
        va_loss = va_sum / n_va

        train_hist.append(tr_loss)
        val_hist.append(va_loss)

        sched.step(va_loss)

        if epoch == 1 or epoch % log_every == 0:
            print(f"[AE] Ep {epoch:4d}  Train {tr_loss:.4e}  Val {va_loss:.4e}")

        if va_loss < best_val - 1e-12:
            best_val = va_loss
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}
            no_imp = 0
        else:
            no_imp += 1
            if no_imp >= patience:
                print(f"[AE] early stop at epoch {epoch}")
                break

    model.load_state_dict(best_state)
    model.to(device)
    model.eval()

    history = {
        "train": np.array(train_hist),
        "val": np.array(val_hist),
    }
    stats = {
        "mu": mu,
        "std": std,
        "val_mse_best": float(best_val),
    }
    return model, history, stats

###############################################################################
#                         4) GP, EI and BO utilities                          #
###############################################################################

def make_gp(latent_dim):
    kernel = 1.0 * Matern(
        length_scale=np.ones(latent_dim),
        length_scale_bounds=(1e-2, 1e2),
        nu=2.5,
    )
    kernel += WhiteKernel(noise_level=1e-6, noise_level_bounds=(1e-8, 1e-2))

    return GaussianProcessRegressor(
        kernel=kernel,
        n_restarts_optimizer=5,
        alpha=1e-6,
        normalize_y=True,
    )

def expected_improvement(x, gp, best_y, xi):
    x = np.atleast_2d(x)
    mu, sigma = gp.predict(x, return_std=True)
    sigma = sigma + 1e-9
    imp = best_y - mu
    z = imp / sigma
    ei = imp * norm.cdf(z) + sigma * norm.pdf(z)
    return ei.ravel()

def propose_sobol_local(
    gp, best_y, center, radius, global_bounds, latent_dim, xi, n_candidates=512
):
    lo_g = global_bounds[:, 0]
    hi_g = global_bounds[:, 1]

    lo = np.maximum(center - radius, lo_g)
    hi = np.minimum(center + radius, hi_g)

    span = hi - lo
    span[span < 1e-6] = 1e-6
    lo = center - 0.5 * span
    hi = center + 0.5 * span

    sampler = qmc.Sobol(d=latent_dim, scramble=True)
    U = sampler.random(n_candidates)
    Zc = qmc.scale(U, l_bounds=lo, u_bounds=hi)

    ei = expected_improvement(Zc, gp, best_y, xi)
    idx = np.argmax(ei)
    return Zc[idx]

def optimise_latent_BO(
    latent_dim,
    objective,
    global_bounds,
    n_initial=40,
    n_iter=40,
    n_candidates=512,
    init_radius=1.0,
    min_radius=0.1,
    seed=42,
):
    rng = np.random.RandomState(seed)

    lo = global_bounds[:, 0]
    hi = global_bounds[:, 1]

    sampler = qmc.Sobol(d=latent_dim, scramble=True)
    U0 = sampler.random(n_initial)
    Z = qmc.scale(U0, l_bounds=lo, u_bounds=hi)

    y = np.array([objective(z) for z in Z])

    best_idx = np.argmin(y)
    best_z = Z[best_idx].copy()
    best_f = y[best_idx]
    radius = init_radius

    gp = make_gp(latent_dim)
    history = [best_f]

    full_span = hi - lo

    for it in range(n_iter):
        gp.fit(Z, y)
        best_y = y.min()

        frac = it / max(1, n_iter - 1)
        xi = 0.1 * (1.0 - frac) + 0.01 * frac

        z_new = propose_sobol_local(
            gp=gp,
            best_y=best_y,
            center=best_z,
            radius=radius,
            global_bounds=global_bounds,
            latent_dim=latent_dim,
            xi=xi,
            n_candidates=n_candidates,
        )

        z_new = z_new + 1e-4 * rng.randn(latent_dim)

        y_new = objective(z_new)

        Z = np.vstack((Z, z_new))
        y = np.append(y, y_new)

        if y_new < best_f:
            best_f = y_new
            best_z = z_new
            radius = min(radius * 1.1, full_span.max() / 2.0)
        else:
            radius = max(radius * 0.9, min_radius)

        history.append(best_f)
        print(f"[BO] it {it+1:02d} | best {best_f:.4e} | radius {radius:.3f}")

    return best_z, best_f, history, Z, y

###############################################################################
#                  5) Latent components for BO for AE                         #
###############################################################################

def build_latent_bo_components(model, mu_stats, std_stats, X, lb, ub, label="AE"):
    print(f"\nBuilding latent space for {label}")

    with torch.no_grad():
        Xn = ((X - mu_stats) / std_stats).astype(np.float32)
        Z_full = model.encode_mu(torch.from_numpy(Xn).to(device)).cpu().numpy()

    scZ = StandardScaler().fit(Z_full)
    Z_wh = scZ.transform(Z_full)

    latent_dim = Z_wh.shape[1]
    z_lo = np.quantile(Z_wh, 0.01, axis=0)
    z_hi = np.quantile(Z_wh, 0.99, axis=0)
    global_bounds = np.stack([z_lo, z_hi], axis=1)

    @torch.no_grad()
    def decode_from_whitened(z_wh_np):
        z = scZ.inverse_transform(np.asarray(z_wh_np, dtype=np.float32)[None, :])
        z_t = torch.from_numpy(z.astype(np.float32)).to(device)
        xhat_norm = model.decode(z_t).cpu().numpy()
        xhat = xhat_norm * std_stats + mu_stats
        xhat = np.clip(xhat[0], lb, ub)
        return xhat

    def latent_objective(z_wh):
        x_dec = decode_from_whitened(z_wh)
        return float(t_func.eval(x_dec))

    components = {
        "latent_dim": latent_dim,
        "global_bounds": global_bounds,
        "decode_fn": decode_from_whitened,
        "objective": latent_objective,
        "scaler": scZ,
    }
    return components

###############################################################################
#                         6) Optuna objective for AE                          #
###############################################################################

def optuna_objective_ae(trial: optuna.Trial):
    t0 = time.perf_counter()

    width_mult = trial.suggest_categorical("width_mult", [0.6, 0.8, 1.0])
    latent_dim = trial.suggest_categorical("latent_dim", [5])
    lr = trial.suggest_float("lr", 1e-4, 3e-3, log=True)
    batch = trial.suggest_categorical("batch", [128, 256, 512])

    print(
        f"\n[Optuna AE] width_mult={width_mult}, latent_dim={latent_dim}, "
        f"lr={lr:.2e}, batch={batch}"
    )

    model, history, stats = train_ae(
        X_full=X,
        latent_dim=latent_dim,
        width_mult=width_mult,
        lr=lr,
        batch_size=batch,
        max_epochs=400,
        patience=60,
        seed=SEED,
        log_every=100,
    )

    val_mse = stats["val_mse_best"]

    trial.set_user_attr("training_time", time.perf_counter() - t0)
    return val_mse

###############################################################################
#                         7) Run Optuna, retrain, BO                          #
###############################################################################

if __name__ == "__main__":
    study = optuna.create_study(direction="minimize")
    study.optimize(optuna_objective_ae, n_trials=10)

    print("\nBest trial:")
    print("  value:", study.best_value)
    print("  params:", study.best_trial.params)

    best_params = study.best_trial.params

    print("\nRetraining AE with best hyperparameters")

    ae_model, ae_hist, ae_stats = train_ae(
        X_full=X,
        latent_dim=best_params["latent_dim"],
        width_mult=best_params["width_mult"],
        lr=best_params["lr"],
        batch_size=best_params["batch"],
        max_epochs=600,
        patience=80,
        seed=SEED,
        log_every=50,
    )

    print("\nAE best val MSE:", ae_stats["val_mse_best"])

    plt.figure(figsize=(6, 4))
    plt.plot(ae_hist["train"], label="AE train")
    plt.plot(ae_hist["val"], label="AE val")
    plt.xlabel("epoch")
    plt.ylabel("loss")
    plt.title("AE training and validation loss")
    plt.legend()
    plt.tight_layout()
    plt.show()

    ae_components = build_latent_bo_components(
        model=ae_model,
        mu_stats=ae_stats["mu"],
        std_stats=ae_stats["std"],
        X=X,
        lb=lb,
        ub=ub,
        label="AE tuned",
    )

    best_z_ae, best_f_ae, hist_ae, Z_all_ae, y_all_ae = optimise_latent_BO(
        latent_dim=ae_components["latent_dim"],
        objective=ae_components["objective"],
        global_bounds=ae_components["global_bounds"],
        n_initial=40,
        n_iter=100,
        n_candidates=512,
        init_radius=1.0,
        min_radius=0.1,
        seed=SEED,
    )

    print("\n[AE latent BO]")
    print("Best latent z:", best_z_ae)
    print("Best objective value:", best_f_ae)
    best_x_ae = ae_components["decode_fn"](best_z_ae)
    print("Decoded best x (first three dims):", best_x_ae[:3])

    plt.figure(figsize=(6, 4))
    plt.plot(hist_ae, label="AE latent BO")
    plt.xlabel("iteration")
    plt.ylabel("best objective so far")
    plt.title("BO in AE latent space")
    plt.legend()
    plt.tight_layout()
    plt.show()
```
